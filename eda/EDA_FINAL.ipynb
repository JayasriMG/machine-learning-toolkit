```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## FILE READING 

```python
def readfile():
    df=pd.read_csv("C:/Users/DHARSHINI/Downloads/diabetes_dataset.csv")
    df.columns=df.columns.str.strip()
    df=df.reset_index(drop=True)
    print(df.head())
    data_inspect(df)
```

## DATA STRUCTURE INSPECT 

```python
def data_inspect_visual(df):
    cols=df.select_dtypes(include=['int64','float64']).columns
    for i in cols:
        plt.figure(figsize=(5,5))
        df[i].hist(bins=20)
        plt.xlabel(i)
        plt.ylabel("Frequency")
        plt.show()
    return 
```

```python
def data_inspect(df):
    print("SIZE:",df.size)
    print("\n\nCOLUMNS\n ",df.columns)
    print("\n\nDATA TYPES\n ",df.dtypes)
    print("\n\nDESCRIBE\n",df.describe())
    print("\n\nINFO\n",df.info())
    data_inspect_visual(df)
    missing_values(df)
    return
```

## MISSING VALUES 

```python
def missingvalues_heat_map(df):
    plt.figure(figsize=(5,5))
    sns.heatmap(df.isnull(),cbar=False,cmap="viridis")
    plt.title("\MISSING VALUES ")
    plt.show()
    return 
    
```

```python
def missing_values(df):
    print("\n\nNULL VALUES",df.isnull().sum())
    print("\n\nNULL VALUES PERCENTAGE",(df.isnull().sum()/len(df))*100)

    missingvalues_heat_map(df)
    duplicate_values(df)
    return 
```

## DUPLICATE VALUES

```python
def duplicate_values(df):
    print("\n\nDUPLICATES VALUES ",df.duplicated().sum())
    categorical_features(df)
    return 
```

## CATEGORICAL FEATURE PROFILING 

```python
def value_count_plot(df):
    plt.figure(figsize=(10,10))
    sns.countplot(data=df)
    plt.title("VALUE COUNT PLOT ")
    plt.show()

    return
```

```python
def categorical_features(df):
    cate=df.select_dtypes(include='object').columns

    for col in cate:
        print("\n\nUNIQUE VALUES :", df[col].unique())
        print("\n\nVALUE COUNT PERCENTAGE ",df[col].value_count())
    value_count_plot(df)
    numerical_feature_analysis(df)
    return 
```

## NUMERICAL FEATURE PROFILING 

```python
def numerical_feature_analysis(df):
    cols=df.select_dtypes(include=['int64','float64']).columns
    print("\n\nDESCRIBE\n",df.describe())

    print("\n\nSKEWNESS")
    for i in cols:
        print(f"{i}:{df[i].skew()}")
    linear_correlation(df)
    return 
```

## MULTIVARIATES ANALYSIS

```python
def linear_correlation(df):
    plt.figure(figsize=(5,5))
    sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm')
    plt.title("LINEAR CORRELATION")
    plt.show()
    outliers(df)
    return 
```

## OUTLIERS VISUALS

```python
def outliers(df):
    print("OUTLIERS VISUALS")
    cate=df.select_dtypes(include=['int64','float64']).columns
    for col in cate:
        plt.figure(figsize=(5,5))
        plt.boxplot(x=df[col])
        plt.xlabel(col)
    return 
```
